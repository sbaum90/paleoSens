```python
"""
This notebook plots the results from the runs in the "ExamplesAndTests_Wimp_reach" folder. 
1) turning on only one background at a time, and varying the exposure,
2) turning on multiple backgrounds,
3) all backgrounds, turning external constraints on the nuisance parameters on/off,
4) 
"""
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set up some plotting settings
fs = 14
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
plt.rcParams.update({'legend.fontsize': fs,
                     'axes.labelsize': fs,
                     'axes.titlesize': fs,
                     'xtick.labelsize': fs,
                     'ytick.labelsize': fs})
plt.rcParams["figure.figsize"] = (9,7)

plt_colors = ['k',
              '#1b9e77',
              '#d95f02',
              '#7570b3',
              '#e7298a',
              '#66a61e',
              '#e6ab02',
              '#a6761d',
              '#666666']
```

```python
# single background, no external constraints
# solar neutrinos, exclusion
fi_list = ['test_single_bkg_nu_solar_2_Gypsum',
          'test_single_bkg_nu_solar_3_Gypsum',
          'test_single_bkg_nu_solar_4_Gypsum',
          'test_single_bkg_nu_solar_5_Gypsum',
          'test_single_bkg_nu_solar_6_Gypsum',
          'test_single_bkg_nu_solar_7_Gypsum',
          'test_single_bkg_nu_solar_8_Gypsum',
          'test_single_bkg_nu_solar_9_Gypsum',
          'test_single_bkg_nu_solar_10_Gypsum']

leg_list = [r'$M = 10^{-2}\,$kg',
           r'$M = 10^{-3}\,$kg',
           r'$M = 10^{-4}\,$kg',
           r'$M = 10^{-5}\,$kg',
           r'$M = 10^{-6}\,$kg',
           r'$M = 10^{-7}\,$kg',
           r'$M = 10^{-8}\,$kg',
           r'$M = 10^{-9}\,$kg',
           r'$M = 10^{-10}\,$kg']

plt.close('all')
for i, fi in enumerate(fi_list):
    in_dat = np.loadtxt(fi+'_exclusion.txt')
    plt.plot(in_dat[:,0], in_dat[:,1], color=plt_colors[i], label=leg_list[i])

plt.xscale('log')
plt.yscale('log')
plt.xlabel(r'$m_\chi$ [GeV]')
plt.ylabel(r'$\sigma_p^{\rm SI}$ [cm$^2$]')
plt.tick_params(right=True,top=True)
plt.tick_params(which='minor',right=True,top=True)
plt.legend()
plt.title(r'Gypsum; exclusion; solar $\nu$ only', y=1.02)
plt.tight_layout()
plt.show()
```
